```python
import numpy as np
```

```python
# buat vector i dengan elemen (1,0) dan vector j dengan elemen (0,1)
i = np.array([1, 0])
j = np.array([0, 1])
```

```python
# buktikan bahwa vector i dan vector j itu orthogonal
dot_ij = np.dot(i, j)
```

```python
# vector v=(1,2) dan w=(2,-1)
v = np.array([1, 2])
w = np.array([2, -1])
```

```python
# cek apakah v dan w orthogonal
dot_vw = np.dot(v, w)

print("Dot product i . j =", dot_ij)
print("Orthogonal (i, j):", dot_ij == 0)

print("\nDot product v . w =", dot_vw)
print("Orthogonal (v, w):", dot_vw == 0)
```
