3. lakukan soal nomor 2 menggunakan Pytorch dan Tensorflow

Pytorch

```
import torch as pt
```

```
x = pt.tensor([[2],
               [1],
               [1]])
x
```

```
x_t = x.T
x_t
```

```
x_t.shape
```

```
y = pt.tensor([[2],
               [1],
               [1]])
y
```

```
y_t = y.T
y_t
```

Tensorflow

```
import tensorflow as tf
```

```
x = tf.constant([[1],
                 [1],
                 [2]])
x
```

```
x_t = tf.transpose(x)
x_t
```

```
y = tf.constant([[1],
                 [1],
                 [2]])
y
```

```
y_t = tf.transpose(y)
y_t
```
